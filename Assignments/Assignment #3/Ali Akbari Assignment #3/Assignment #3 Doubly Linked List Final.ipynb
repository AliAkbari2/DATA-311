```python
#Inner class node to be used for doubly linked list 
#Constuctor takes value of node & pointer of next and previous values else they are defaulted to None
class Node:
    def __init__(self, val, nextPointer = None, previousPointer = None):
        self.val = val
        self.nextPointer = nextPointer
        self.previousPointer = previousPointer  
```
