```python
def ifValid(exp):
    counter = 0
    binOperatorsList = ["+","-","*","/"]
    for i in range(len(exp)):
        if counter < 0:
                return -1
        elif exp[i] == " ":
            continue
        elif type(exp[i]) is str:
            try:
                x = float(exp[i])
                counter += 1
            except:
                #pass
                if exp[i] in binOperatorsList:
                    counter = counter - 2
                    if counter < 0:
                        return -1
                    counter += 1
                else: 
                    return -1
   
    if counter == 1:
        return 1

def applyOperation(aStr, aFloat, aFloat2):
    if aStr == '+':
        return aFloat + aFloat2
    elif aStr == '-':
        return aFloat2 - aFloat
    elif aStr == '*':
        return aFloat * aFloat2
    elif aStr == '/':
        return aFloat2 / aFloat    
```
